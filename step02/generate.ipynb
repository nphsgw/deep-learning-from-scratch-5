```python
import numpy as np
import matplotlib.pyplot as plt

path = "./height.txt"
xs = np.loadtxt(path)
# パラメータ推定
mu = np.mean(xs)
sigma = np.std(xs)

# 作成した生成モデルを使ってデータを生成して、パラメータ推定に使ったサンプルデータと比べてみる。
# 指定した平均と標準偏差の正規分布から10000個のデータを生成
samples = np.random.normal(mu, sigma, 10000)


plt.hist(xs, bins='auto', density=True, alpha=0.7, label='original')
plt.hist(samples, bins='auto', density=True, alpha=0.7, label='generated')
plt.xlabel('Height(cm)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

パラメータθによって形状が決まる確率分布があるとする\
サンプルD（確率変数の塊）が得られ、各データがある確率分布p(D;θ)に従って独立に得られるとする。\
p(D;θ)はパラメータをθとした場合にサンプルDが得られる確率密度を表す。\
このp(D;θ)をθを引数とする関数とみなすと次の式で定義することもできる。\
L(θ)=p(D;θ)\
このL(θ)を尤度や尤度関数と呼ぶ。\
尤度はパラメータθを引数とする関数であり、あるパラメータθに対してサンプルDがどれぐらい起こり得るかを確率密度として表す。\

最尤推定は、尤度p(D;θ)を最大化するパラメータθを見つける手法である。\


・生成モデル\
　あるデータxの確率分布p(x)をモデル化（数式で表す）ことで、あたかもその集団から選んだような擬似的なデータを新たに生成できるモデルのこと。\
　生成モデルを作るには、モデル化とパラメータ推定を行うのが一般的。\
・モデル化\
  真の確率分布がパラメータで調整可能な確率分布で近似されると仮定する\
・パラメータ推定\
  その確率分布がサンプルデータに合うようにパラメータを推定する。\

今回のデータセットは正規分布のような形をしていたので、
モデル化＝母集団も正規分布だと仮定\
パラメータ推定＝ここでは最尤推定でサンプルに最もフィットするパラメータに調整する\

パラメータ推定について\
正規分布のパラメータは平均と標準偏差。\
このパラメータを最尤推定によって求める。\
つまり、最尤推定により正規分布の確率密度関数が最大になるパラメータを求めることである。\
正規分布の確率密度関数の式を微分すると、正規分布の平均と標準偏差を算出する式に帰結する。\
よって、正規分布を仮定した生成モデルのパラメータを最尤推定で求めると、サンプルから平均と標準偏差を求めることになる。
